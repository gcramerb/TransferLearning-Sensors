```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
actLabels = ["Walking", "Ascending\n stairs", "Descending\n stairs","Lying","Sitting","Standing"]
```

### MultiSource -> DSADS (6 Atividades)

```python
arrayDsa = np.array([[353, 240, 252,   3,   0,   0],
       [ 65, 202,  90,  30,   0,   1],
       [ 61,  38, 133,  21,  57,   0],
       [  0,   0,   0, 723, 238,   0],
       [  1,   0,   5, 103, 185,   0],
       [  0,   0,   0,  80,   0, 479]]).transpose()
df_cm = pd.DataFrame(arrayDsa, index = actLabels,
                  columns = actLabels)
plt.figure(figsize = (10,8))
plt.title("MultiSource to DSADS",fontsize=20)

sn.heatmap(df_cm/np.sum(df_cm), annot=True, 
            fmt='.2%', cmap='Blues')
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0, ha="right")
plt.xlabel("True Labels",fontsize=14)
plt.ylabel("Predicted Labels",fontsize=14)
plt.savefig('C:\\Users\\gcram\\Desktop\\confusionMatrixDissertação\\multiSourceDsads.pdf')
```
