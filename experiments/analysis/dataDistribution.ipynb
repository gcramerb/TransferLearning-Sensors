```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
num_points = 120
point_size = 200
```

```python
mean1 = np.array([5,2])
cov1 = np.array([[0.7,1],[1,0.7]])
mean2 = np.array([2.1,2])
cov2 = np.array([[1,-0.2],[-0.2,1]])

#sns.set_style("white")
points1 = np.random.multivariate_normal(mean1, cov1, num_points)
pointsDF1 = pd.DataFrame(points1,columns=["x","y"])
pointsDF1["Class"] = 1

points2 = np.random.multivariate_normal(mean2, cov2, num_points)
pointsDF2 = pd.DataFrame(points2,columns=["x","y"])
pointsDF2["Class"] =2
points = np.concatenate([points1,points2])
df = pd.concat([pointsDF1,pointsDF2 ],axis = 0)
sns.kdeplot(x=points[:,0], y=points[:,1],levels=4, thresh=.2,color ="red")
sns.scatterplot(data=df, x = "x", y = "y",s = [point_size],  style="Class",c = "red")
plt.show()
```

```python
mean1 = np.array([5,2])
cov1 = np.array([[1,-0.5],[-0.5,1]])
mean2 = np.array([5,2])
cov2 = np.array([[1,-0.7],[-0.7,1]])
points1 = np.random.multivariate_normal(mean1, cov1, num_points)
pointsDF1 = pd.DataFrame(points1,columns=["x","y"])
pointsDF1["Class"] = 1

points2 = np.random.multivariate_normal(mean2, cov2, num_points)
pointsDF2 = pd.DataFrame(points2,columns=["x","y"])
pointsDF2["Class"] =2

points = np.concatenate([points1,points2])
df = pd.concat([pointsDF1,pointsDF2 ],axis = 0)
sns.kdeplot(x=points[:,0], y=points[:,1],levels=4, thresh=.2,color ="blue")
sns.scatterplot(data=df, x = "x", y = "y",s = [point_size],  style="Class", c = "blue")
plt.show()
```
